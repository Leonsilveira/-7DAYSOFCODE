```python
import pandas as pd

# Lista de arquivos CSV
arquivos = [
    r"E:\Projetos 2025\#7DaysOfCode\dataset\despesa_ceaps_2019.csv",
    r"E:\Projetos 2025\#7DaysOfCode\dataset\despesa_ceaps_2020.csv",
    r"E:\Projetos 2025\#7DaysOfCode\dataset\despesa_ceaps_2021.csv",
    r"E:\Projetos 2025\#7DaysOfCode\dataset\despesa_ceaps_2022.csv"
]

# Criar lista de DataFrames
df_list = [pd.read_csv(arquivo, delimiter=';', encoding='ISO-8859-1') for arquivo in arquivos]

# Concatenar os DataFrames em um só
df_ceaps = pd.concat(df_list, ignore_index=True)

# Exibir as primeiras linhas para conferir
print(df_ceaps.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados (substituir pelo caminho do seu arquivo, se necessário)
df = pd.read_csv(r'E:\Projetos 2025\#7DaysOfCode\dataset\df_ceaps_limpo.csv')  # Substitua pelo caminho correto

# 1️⃣ Estatísticas básicas - Total de gastos por senador
gastos_por_senador = df.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
print("Total de gastos por senador:")
print(gastos_por_senador.head(10))  # Exibe os 10 maiores

# Gráfico de barras - Total de gastos por senador
plt.figure(figsize=(12, 6))
gastos_por_senador.plot(kind='bar', color='skyblue')
plt.title('Total de Gastos por Senador')
plt.xlabel('Senador')
plt.ylabel('Valor Reembolsado (R$)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 2️⃣ Visualização gráfica - Top 10 senadores com maiores gastos
plt.figure(figsize=(12, 6))
gastos_por_senador.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Senadores com Maiores Gastos')
plt.xlabel('Senador')
plt.ylabel('Valor Reembolsado (R$)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Gráfico de pizza - Distribuição dos gastos dos Top 10 Senadores
plt.figure(figsize=(10, 6))
gastos_por_senador.head(10).plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribuição dos Gastos dos Top 10 Senadores')
plt.ylabel('')
plt.show()
```

```python
# 3️⃣ Consulta específica - Filtrar despesas de um senador específico
senador_escolhido = 'ACIR GURGACZ'  # Alterar conforme necessário
filtro_senador = df[df['SENADOR'] == senador_escolhido]
print(f"Despesas do senador {senador_escolhido}:")
print(filtro_senador[['TIPO_DESPESA', 'FORNECEDOR', 'VALOR_REEMBOLSADO']].head(10))

# Gráfico de barras - Despesas de um senador específico por tipo de despesa
plt.figure(figsize=(12, 6))
gastos_por_tipo = filtro_senador.groupby('TIPO_DESPESA')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
gastos_por_tipo.plot(kind='bar', color='skyblue')
plt.title(f'Despesas por Tipo de Despesa - {senador_escolhido}')
plt.xlabel('Tipo de Despesa')
plt.ylabel('Valor Reembolsado (R$)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Gráfico de pizza - Distribuição das Despesas de um Senador Específico por Tipo de Despesa
plt.figure(figsize=(10, 6))
gastos_por_tipo.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title(f'Distribuição das Despesas por Tipo de Despesa - {senador_escolhido}')
plt.ylabel('')
plt.show()
```
